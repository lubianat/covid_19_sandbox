```python
import pandas as pd


viral_proteins_df = pd.read_csv("SARS-CoV-2 Manually Curated from PDB - Sheet1.csv")
```

```python
viral_proteins_df = viral_proteins_df.sort_values("label")
viral_proteins_df["label"] = viral_proteins_df["label"].str.strip()
viral_proteins_df
```

```python
# create pages
current_label = ""
for index, row in viral_proteins_df.iterrows():
    if row["label"] != current_label:
        print("""
        CREATE
        LAST|Len|""" +  row["label"] + """
        LAST|Den|""" +  row["description"] + """
        LAST|P31|""" +  row["P31"]  + """
        LAST|P279|""" + row["P279"]  + """
        LAST|P703|""" + row["P703"]  + """
        LAST|P638|""" + '"' + row["P638"] + '"'  + """
        """)
      
    if row["label"] == current_label:
        print( "        LAST|P638| " + row["P638"]  )
        
  
    current_label = row["label"]
```
